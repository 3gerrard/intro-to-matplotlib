```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
student_details = {
    "names": ["asheley" ,"gerrard " ,"kevin" ,"aiden" ,"aron" ,"bismark" ,"dany" ,"lucy" ,"blessing" ,"angel" ],
    "scores":[23,100,23,45,6,7,89,9,64,44,]
}
print(student_details)

```

```python
df = pd.DataFrame(student_details)
df.sort_values(by="scores" , ascending=False, inplace=True)

```

```python
plt.plot(df["names"], df['scores'])
plt.xlabel('names')
plt.ylabel('scores')
plt.title('student_scores')
plt.show()
```

```python
plt.bar(df["names"], df['scores'])
plt.xlabel('names')
plt.ylabel('scores')
plt.xticks(rotation=45)
plt.title('student_scores')
plt.show()
```

* gerrard is the highest performing student,because he has thw longest bar.
* aron is the lowest performing student because he ahs the shortest bar.
